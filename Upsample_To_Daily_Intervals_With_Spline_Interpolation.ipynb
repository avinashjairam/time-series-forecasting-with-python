```python
#importing libraries
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot as plt
```

```python
#parser
def parser(x):
  return datetime.strptime('190' + x, '%Y-%m')
```

```python
#reading in the data from the csv file
series = read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv', header=0, index_col=0, parse_dates=True, date_parser=parser).squeeze('columns')
```

```python
#upsampling
upsample = series.resample('D').mean()
```

```python
#interpolation
interpolated = upsample.interpolate(method='spline', order=2)
```

```python
#printing the interpolated values
print(interpolated.head(32))
```

```python
#plotting the interpolation
interpolated.plot()
plt.show()
```
