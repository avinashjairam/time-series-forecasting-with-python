```python
#create a heat map of yearly data
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
import matplotlib.pyplot as plt
```

```python
#reading in data in the form of a series 
series = read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True).squeeze('columns')
```

```python
#creating a group
groups = series.groupby(Grouper(freq='A'))
```

```python
#creating a dataframe
years = DataFrame()
```

```python
#extracting name and group from groups
for name, group in groups:
  years[name.year] = group.values
```

```python
years
```

```python
#transpose of year
years = years.T
```

```python
#plotting the graph
plt.matshow(years, interpolation=None, aspect='auto')
plt.show()
```
