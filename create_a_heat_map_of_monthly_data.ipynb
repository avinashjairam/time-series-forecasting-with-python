```python
#create a heat map of monthly data
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
import matplotlib.pyplot as plt
from pandas import concat
```

```python
#reading in data in the form of a series 
series = read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True).squeeze('columns')
```

```python
#selecting the year 1990
one_year = series['1990']
```

```python
#creating a group
groups = one_year.groupby(Grouper(freq='M'))
```

```python
#joining the months
months = concat([DataFrame(x[1].values) for x in groups], axis = 1)
```

```python
#creating a dataframe of months
months = DataFrame(months)
```

```python
#setting numerical columns to correspond to the months
months.columns = range(1,13)
```

```python
#plotting the graph
plt.matshow(months, interpolation=None, aspect='auto')
plt.show()
```
